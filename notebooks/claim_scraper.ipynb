```python
import urllib.request
from bs4 import BeautifulSoup as bs
import pandas as pd
from datetime import datetime, timedelta
```

```python
df = pd.DataFrame(columns=['veracity', 'Claim_by', 'Claim'])
```

```python
day = datetime.today()
for _ in range(365):
  link =  'https://mediabiasfactcheck.com/' + day.strftime('%Y/%m/%d') + '/the-latest-fact-checks-curated-by-media-bias-fact-check-' + day.strftime('%m-%d-%Y')
  day = day - timedelta(days=1)

  try:
    page = urllib.request.urlopen(link)
  except:
    continue
  soup = bs(page, 'html.parser')

  rows = soup.find_all('tr')
  d = {}
  for row in rows:
    d['veracity'] = row.find_all('td')[0].text.replace('\n', ' ')
    claim_info =row.find_all('td')[1].text
    d['Claim_by'] = claim_info.split(':')[0]
    d['Claim'] = claim_info.split(':')[-1]
    #d = pd.DataFrame(d, index=[0])
    df = df.append(d, ignore_index=True)
```
