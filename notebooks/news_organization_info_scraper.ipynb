```python
import urllib.request
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
link = 'https://mediabiasfactcheck.com/filtered-search/?pg=1'
page = urllib.request.urlopen(link)
soup = bs(page, 'html.parser')
```

```python
# Defining of the dataframe
df = pd.DataFrame(columns=['News_Source',	'Bias', 'Reporting',	'Country',	'Credibility',	'Media_Type',	'Traffic', 'Popularity_Estimate', 'link'])
for i in range(44):
  link = 'https://mediabiasfactcheck.com/filtered-search/?pg=' + str(i+1)
  page = urllib.request.urlopen(link)
  soup = bs(page, 'html.parser')

  table =  soup.find('table', class_="mbfc-table")
  # Collecting Ddata
  for row in table.tbody.find_all('tr'):    
      # Find all data for each column
      columns = row.find_all('td')
      
      if(columns != []):
          News_Source = columns[0].text.strip()
          Bias = columns[1].text.strip()
          Reporting = columns[2].text.strip()
          Country = columns[3].text.strip()
          Credibility = columns[4].text.strip()
          Media_Type = columns[5].text.strip()
          Traffic = columns[6].text.strip()
          Popularity_Estimate = columns[6].text.strip()
          link = row.a['href']
          


          df = df.append({'News_Source':News_Source,	'Bias':Bias, 'Reporting':Reporting,	'Country':Country,	'Credibility':Credibility,	'Media_Type':Media_Type,	'Traffic':Traffic, 'Popularity_Estimate':Popularity_Estimate, 'link':link}, ignore_index=True)
```
